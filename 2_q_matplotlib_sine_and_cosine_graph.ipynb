```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = [0, 30, 45, 60, 90]
x = [i*(np.pi/180) for i in t]

plt.plot(t, np.sin(x), 'ro:', markersize = 8, markerfacecolor = 'k', markeredgecolor = 'k', label = 'sin')
plt.plot(t, np.cos(x), 'k:', marker= 'o', markerfacecolor = 'r', markeredgecolor = 'r', markersize = 8, label = 'cos')

plt.xlabel('Angle')
plt.ylabel('sin and cos value')

plt.xticks(t, backgroundcolor = 'y')

plt.minorticks_on()
plt.grid(b = True, which = "major", color = 'purple')
plt.grid(b = True, which = 'minor')

plt.legend(loc = 'center left', edgecolor = 'r', fontsize = '10', title = 'Angle', title_fontsize = '12',
          facecolor = 'y', framealpha = 0.2)

plt.show()
```

```python
t = np.linspace(-np.pi, np.pi)

plt.plot(t, np.sin(t), 'ro:',  markerfacecolor = 'k', markeredgecolor = 'k')
plt.plot(t, np.cos(t), 'ko:', markerfacecolor = 'r', markeredgecolor = 'r')


plt.xlabel('Angle', fontsize = '14')
plt.ylabel('Value', fontsize = '14')
plt.title('sin & cos graph', fontsize = '18')

plt.xticks( rotation = 45)

plt.show()
```

```python
t = np.arange(0, 4*np.pi, 0.5)
plt.plot(t, np.sin(t), 'ro:')
plt.plot(t, np.cos(t), 'yo:')

plt.show()
```
