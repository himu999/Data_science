```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("E:/Data_science/gas_prices.csv")


plt.figure(figsize=(8, 5), dpi=100)

plt.title('Gas Prices over Time(in USD)', fontdict={'fontweight':'bold', 'fontsize':18})




plt.plot(df.Year, df.USA, 'b.-', label='USA')
plt.plot(df.Year, df.Canada, 'r.-', label = 'Canada' )
plt.plot(df.Year, df['South Korea'], 'g.-', label='South Korea')
plt.plot(df.Year, df.Australia, 'y.-', label='Australia')


# another way to plot many values
# country_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in df:
#     if country in country_look_at:
#         plt.plot(df.Year, df[country], marker=".",  label = country)
# for country in df:
#     if country != "Year":
#         plt.plot(df.Year, df[country])
    
    
    
plt.xticks(df.Year[::2].tolist()+[2011], color='b')
plt.yticks(color='b')

plt.xlabel('Year', fontdict={'fontweight':'bold', 'fontsize':'14'}, color="r")
plt.ylabel('US Dollars', color='r', fontweight='bold', fontsize="14")


plt.legend()
plt.savefig('gas_price_figure1', dpi=300)
plt.show()
```
