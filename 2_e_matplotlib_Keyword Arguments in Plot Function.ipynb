```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
x = [1, 2, 3]

ax = plt.figure().gca()
ax.xaxis.set_major_locator(MaxNLocator(integer = True))
ax.yaxis.set_major_locator(MaxNLocator(integer = True))

plt.plot(x, 'r:o')

plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Simple Plot')

plt.show()
```

```python
x = np.arange(0, 10+0.5, 0.5)
plt.plot(x, 'go:', x**2.5, 'r-.o', x**3, 'k:o')
```

# Antialiased

```python
x = [10, 20, 30, 40]

# plt.plot(x, linewidth=20, antialiased = True)
plt.plot(x, 'r:', linewidth = 20, antialiased = False, )

plt.show()
```

# dashes

```python
x = [10, 20, 30]
plt.plot(x, dashes=[3, 10])
plt.plot([12, 3, 4], dashes=[1, 2])
plt.plot([5, 6, 7, 8, 9], dashes=[3, 15])
plt.show()
```

# linestyle or ls

```python
plt.plot([10, 20, 30, 40], linestyle = "dotted")
plt.show()
```

# marker edge color , marker edge width, markersize

```python
x = [1, 2, 3, 4, 5]
plt.plot(x, ':', marker='.',  markeredgecolor = 'r', markeredgewidth  = '10', markersize ='10', markerfacecolor = 'r')
plt.show()
```

# markevery

```python
x = [1, 2, 3,  4,  5, 6,  7, 8,]
plt.plot(x, 'r:o', markevery=2)
plt.show()
```
