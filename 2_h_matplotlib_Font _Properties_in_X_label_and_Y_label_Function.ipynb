```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3,  4,  5]

plt.plot(x, 'r:o', markeredgecolor = 'k', markerfacecolor = 'k')

plt.xlabel('x label', labelpad = 20)
plt.ylabel('y label', {"backgroundcolor" : 'orange', "color": "k", 'size':20, 'weight':'bold', 'fontfamily':'fantasy'},
           labelpad = 20,)
plt.title('Simple Plot', backgroundcolor = 'r', color= 'blue', weight = 200,  fontfamily='monospace' )
plt.show()
```
