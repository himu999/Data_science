```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]

left, right = plt.xlim()
print(left)
print(right)

plt.plot(x)

plt.show()
```

```python
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.plot([1, 2,  3, 4])

plt.show()
```

```python
plt.ylim(3, 6)

plt.plot([1, 2,  3, 4])

plt.show()
```

```python
plt.xlim(left = 3, right = 6)
plt.ylim(bottom = 3, top = 6)
plt.plot([1, 2, 3, 4])
plt.show()
```
