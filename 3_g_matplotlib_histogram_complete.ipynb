# Histogram complete tutorial                                                                                                                                                                                

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python

x = [1, 21, 12, 13, 22, 28, 27, 9, 14, 36, 31, 39, 40, 12, 16, 7]
y = [1, 6, 11, 16, 21, 26, 31, 36, 41]

ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer=True))

n, bins, patches = plt.hist(x, y, ec='g', color='b')
patches[0].set_fc('r')
patches[1].set_fc('g')
patches[2].set_fc('orange')
patches[3].set_fc('r')
patches[4].set_fc('g')
patches[7].set_fc('y')

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Population Age Count')

plt.xticks(y, color='r')

plt.show()
```

```python
x = [1, 2, 3, 4, 4, 6, 6, 9, 10, 11, 13, 13, 12, 15, 17, 19, 18, 20, 21, 23, 25, 27, 31, 36, 39, 41]
y = [1, 6, 11, 16, 21, 26, 31, 36, 41]

ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer = True))

n, bins, patches = plt.hist(x, y, ec = 'r', orientation = 'horizontal')
color = ['r', 'g', 'b', 'orange', 'crimson', 'r', 'g', 'b']

ln = len(patches)

for i in range(ln):
    patches[i].set_fc(color[i])
    
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Population Age Count')
plt.yticks(y)

plt.show()
```

```python
x = [1, 1, 2, 3, 4, 5, 6, 7, 10, 10, 10, 13, 12, 12, 12, 11, 14, 15, 18, 18, 19, 21, 23, 23, 25, 27, 31,
     31, 31, 33, 37, 37, 39, 39, 40, 40, 41]
y = [1, 6, 11, 16, 21, 26, 31, 36, 41]

ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer = True))

n, bins, patches = plt.hist(x, y, ec = 'r', bottom = [0, 1, 2, 3, 4, 5, 6, 7])
color = ['r', 'orange', 'crimson', 'r', 'orange', 'crimson', 'r', 'orange']

for i in range(len(n)):
    patches[i].set_fc(color[i])

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Population Age Count')
plt.xticks(y)

plt.show()
```
