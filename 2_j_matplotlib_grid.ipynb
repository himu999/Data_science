```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(x, y)

plt.minorticks_on()

plt.grid(b = True, which='minor', color = 'r')
plt.grid(b = True, color = 'g', linewidth = 1)


plt.show()
```

# Axis

```python
from matplotlib.pyplot import MaxNLocator
x = [1, 2, 3, 4]

ax = plt.figure().gca()
ax.xaxis.set_major_locator(MaxNLocator(integer = True))
ax.yaxis.set_major_locator(MaxNLocator(integer = True))

plt.plot(x,x)

plt.minorticks_on()
plt.grid(b = True, which = 'minor', color = 'r', axis = 'x')

plt.show(block = True)

```

```python
x = [1, 2, 3, 4]

plt.plot(x, x)

plt.minorticks_on()

plt.grid(b = True, which = 'minor', color = 'orange', axis = 'y')

plt.show()
```

```python
x = [1, 2, 3, 4]

plt.plot(x, x, marker = 'o',markeredgecolor = 'k', markerfacecolor = 'k')

plt.minorticks_on()

plt.grid(b = True, which = 'minor', axis = 'both', color = 'r')
plt.show()
```
