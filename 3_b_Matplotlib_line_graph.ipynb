# Load necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Line graph

```python
gas = pd.read_csv('E:\Data_science\gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title('Gas prices over time (in USD)')

# plt.plot(gas.Year, gas.USA,'r.-',label="USA",)
# plt.plot(gas.Year, gas.Canada, 'g.-', label="Canada")
# plt.plot(gas.Year, gas['South Korea'],'C1.-', label="South Korea")
# plt.plot(gas.Year, gas.Australia, 'Black', marker=".", markeredgecolor='r', label='Australia')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3], color="b")
plt.yticks(color="b")
plt.xlabel('Year', fontsize=15, color="r")
plt.ylabel('US dollar', fontsize=15, color='r')
plt.legend()

plt.show()
```
