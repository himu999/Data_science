```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
x = np.array([1, 2, 3,  4,])

ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer = True))

plt.plot(x**2, x, 'ro:', markersize= 8, markerfacecolor = 'k', markeredgecolor = 'k', label = 'squares')
plt.plot(x**3, x, 'ko:', markersize = 8, markerfacecolor = 'r', markeredgecolor = 'r', label = 'cubes')

plt.xlabel('x label', size = 14)
plt.ylabel('y label', size = 14)
plt.title('Simple plot', size = 16)

plt.ylim(0, 5)
plt.xlim(0, 70)

plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor', )

# plt.legend(loc = 2, ncol=2)


plt.legend(loc = (0/70, 1/5))

plt.show()
```

# bbox to anchor

```python
x = np.array([1, 2, 3, 4])
plt.plot(x**2, x, 'ro:', markersize = 8, markerfacecolor = 'k', markeredgecolor = 'k', label = 'squares')
plt.plot(x**3, x, 'ko:', markersize = 8, markerfacecolor = 'r', markeredgecolor = 'r', label = 'cubes')
plt.plot(x, x**2, 'bo:', markersize = 8, markerfacecolor = 'orange', markeredgecolor = 'orange', label = 'line3')
plt.plot([0, 1, 2, 3], x**2.2, 'mo:', markersize = 8, markeredgecolor = 'b', markerfacecolor = 'b', label= 'line4')

plt.xlim(0, 70)
plt.ylim(0, 25)

plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')

# plt.legend(loc = 'upper left', bbox_to_anchor=(0.5,0.5,0.5,0.5))
plt.legend(loc = 'lower left', bbox_to_anchor = (0.5, 0.5))
plt.show()
```

# place the legend outside the plot

```python
x = np.array([1, 2, 3, 4])

plt.plot(x**2, x, 'ro:', markeredgecolor = 'k', markerfacecolor = 'k', markersize = 8, label = 'squares')
plt.plot(x**3, x, 'ko:', markeredgecolor = 'r', markerfacecolor = 'r', markersize = '8', label = 'cubes')

plt.xlim(0, 70)
plt.ylim(0, 5)

plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')

# plt.legend(loc = 'upper right', bbox_to_anchor = (0, 0))

# plt.legend(loc = (0.427, 1))
plt.legend(loc = 'upper left', fontsize = 'medium', facecolor = 'y')
plt.tight_layout()

plt.show()
```

# font size and font color

```python
x = np.array([1, 2, 3, 4])

plt.plot(x**2, x, 'go:', markersize = '8', markerfacecolor = 'r', markeredgecolor = 'r', label = 'squares')
plt.plot(x**3, x, 'ro:', markersize = '8', markerfacecolor = 'k', markeredgecolor = 'k', label = 'cubes')

plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')

plt.xlim(0, 70)
plt.ylim(0, 5)

plt.xlabel('x label', fontsize = 14, color = 'r')
plt.ylabel('y label', fontsize = 14, color = 'r')
plt.title('Simple plot', fontsize = 16, color = 'r')

plt.legend(loc = 'upper left', fontsize = 'medium', facecolor ='y', edgecolor = 'r', framealpha=0.2)


plt.show()
```

# changing legend looks

```python
x = np.array([1, 2, 3, 4])

plt.plot(x**2, x, 'ro:', markersize = '8', markerfacecolor = 'k', markeredgecolor  = 'k', label = 'label1')
plt.plot(x**3, x, 'ko:', markersize = '8', markerfacecolor = 'r', markeredgecolor = 'r', label = 'label2')

plt.xlabel('x label', fontsize = 14)
plt.ylabel('y label', fontsize = 14)
plt.title('Simple plot', fontsize = 16)

plt.xlim(0, 70)
plt.ylim(0, 5)

plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'purple')
plt.grid(b = True, which = 'minor')

plt.legend(loc = 'upper left', facecolor = 'y', edgecolor = 'r', ncol=2, framealpha = 0.2, markerfirst = False, 
           frameon = True, columnspacing = 5)

plt.show()
```

# how to give title to the legend

```python
x = np.array([0, 1, 2, 3, 4])

plt.plot(x**2,x, 'ro:', markersize = '8', markerfacecolor = 'k', markeredgecolor = 'k', label = 'label1')
plt.plot(x**3, x, 'ko:', markersize = '8', markerfacecolor = 'r', markeredgecolor = 'r', label = 'label2')

plt.xlim(0, 70)
plt.ylim(0, 5)

plt.xlabel('x label', fontsize = '14')
plt.ylabel('y label', fontsize = '14')
plt.title('Simple plot', fontsize = 16)

plt.legend(loc = 'center right' , title = 'my legend', edgecolor = 'r', title_fontsize = 14, borderpad = 2, shadow=True,
          labelspacing=2)


plt.show()
```

# multiple legend in single plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 2, 3, 4])

l1, = plt.plot(x**2.85, x**1.5, 'ro:', markersize = '8', markerfacecolor = 'k', markeredgecolor = 'k', label = 'label1')
l2, = plt.plot(x**3, x, 'ko:', markersize = '8', markerfacecolor = 'r', markeredgecolor = 'r', label = 'label2')
l3, = plt.plot(y**3, x**2.5, 'bo:', markersize = '8', markerfacecolor = 'purple', markeredgecolor = 'purple',
                  label = 'label3')
l4, = plt.plot(y**3, x**2, 'mo:', markersize = '8', markerfacecolor = 'y', markeredgecolor = 'y', label = 'label4')

plt.xlim(0, 70)
plt.ylim(0, 35)

plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')

legend1 = plt.legend((l1,l2), ['label1','label2'], facecolor = 'y')
plt.gca().add_artist(legend1)
legend2 = plt.legend((l3, l4), ['label3', 'label4'], facecolor = 'g', loc = 'center left')
plt.gca().add_artist(legend2)
plt.show()
```
