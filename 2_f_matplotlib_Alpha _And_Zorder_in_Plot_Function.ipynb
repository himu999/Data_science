```python
import matplotlib.pyplot as plt
```

```python
plt.plot( [1, 2, 3, 4], [5, 6, 7, 8], 'r:o', markerfacecolor = 'k', markeredgecolor='k')
plt.plot([5, 6, 7, 8], 'r:o', markerfacecolor='k', markeredgecolor = 'k', alpha = 0.75)
plt.plot([6, 7, 8, 9], 'r:o', markerfacecolor = 'k', markeredgecolor = 'k', alpha = 0.5)
plt.plot([7, 8, 9, 10], 'r:o', markerfacecolor = 'k', markeredgecolor = 'k', alpha = 0.25)

plt.show()
```

# zorder

```python
plt.plot([10, 20, 80, 90], zorder = 2)
plt.plot([10, 20, 30, 40], zorder = 1)


plt.show()
```

```python
plt.plot([0, 0], [-1, 1], linewidth = 10, zorder = 3)
plt.plot([-1, 1], [-1, 1], linewidth = 10, zorder = 1)
plt.plot([-1, 1], [1, -1], linewidth = 10, zorder = 2)
plt.show()
```
