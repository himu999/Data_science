```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
x = [1, 2, 3, 4]
y = [0, 1, 2, 3]

ax = plt.figure().gca()
ax.xaxis.set_major_locator(MaxNLocator(integer = True))
ax.yaxis.set_major_locator(MaxNLocator(integer = True))

line1, = plt.plot(x, 'ro:', markerfacecolor = 'k', markeredgecolor = 'k', markersize = '8', label = 'First line')
line2, = plt.plot(y, 'ko:', markerfacecolor = 'r', markeredgecolor = 'r', markersize = '8', label = 'Second line')


plt.xlabel('x label', size = '14')
plt.ylabel('y label', size = '14')
plt.title('Simple plot', size = '16')

plt.xlim(0, 6)
plt.ylim(0,6)

plt.minorticks_on()

plt.grid(b= True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')

plt.legend([line2, line1], ['Second line', 'First line'])

plt.show()
```
