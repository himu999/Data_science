# FontDict

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]

plt.plot(x)

plt.xlabel('x label', fontdict =  {'size': 20, "color": "r"}, labelpad = -40)
plt.ylabel('y label', size = '20', color = 'r', labelpad = -50)
plt.title('Simple Plot Graph', size = '25', color = 'r')

plt.show()
```
