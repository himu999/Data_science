# Legend

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
t = np.array([1, 2, 3, 4])

"""
Type - 1
plt.plot(t**2, t, color = 'r', label = 'T squares')
plt.plot(t**3, t, color = 'g', label = 'T cubes')
"""

"""
Type - 2
plt.plot(t**2, t, color = 'r')
plt.plot(t**3, t, color = 'g')
"""


line1, = plt.plot(t**2, t, color = 'r')
line2, = plt.plot(t**3, t, color = 'g')



plt.xlabel('x label')
plt.ylabel('y label')
plt.title('simple plot')


"""
Type - 1
plt.legend()
"""

"""
Type - 2
plt.legend(['squares', 'cubes'])
plt.legend(['', 'cubes'])
"""

# plt.legend([line1, line2], ['squares', 'cubes'])
# plt.legend(['squares', 'cubes'])

plt.legend((line2,), ['cubes'])

# plt.legend([line2, line1], ['cubes', 'squares'])

plt.show()
```
