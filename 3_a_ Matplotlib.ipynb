# Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize your Graph (dpi specifies pixels per inch. when saving probably should use 300 if possible )
plt.figure(figsize=(5, 3), dpi=100)

# Line 1

#keyword argument notation
# plt.plot(x, y, label="2x", color='r', linewidth=2, marker='.', markersize="10", markeredgecolor='b', linestyle='--' )

# use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'r.-', label='2x', markersize='10')

#line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'y', label='x^2')

# plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'b--')



# Add a title (specify font parameters with font dict)
plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('X Axis!' )
plt.ylabel('Y Axis!')

# X, Y axis Tickmarks 
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7, 10])

# Add a legend
plt.legend()

# Save Figure (dpi 300 is good when saving so)
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]
bars = plt.bar(labels, values)
# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(6,4))

plt.show()
```
