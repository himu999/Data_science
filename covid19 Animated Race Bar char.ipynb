```python
os.chdir("E:\Data_science")
data_corona = pd.read_csv("E:\Data_science\corona_data19.csv")
data_corona.head(5)
```

```python
cols = ['date', 'Italy', 'China', 'Australia', 'Brazil', 'India', 'United Kingdom', 'Pakistan', 'Spain', 'Canada', 'New Zealand',
       'Russia', 'Thailand']

subsetdf = data_corona[cols]
subsetdf.set_index('date', inplace = True)
cum_sum_df = subsetdf.cumsum(axis = 0) #cumulative sum of all column
cum_sum_df.tail(10)
```
