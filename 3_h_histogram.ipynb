# Histogram

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
x = [1, 1, 2, 3, 5, 5, 7, 7, 9, 10, 11, 6, 16, 14, 12, 13, 15, 15, 15, 15, 21, 17, 19, 18, 21, 21, 21, 
    23, 25, 25, 21, 27, 29, 29, 29, 30, 31, 36, 35, 3, 33, 33, 33, 33, 34, 36, 36, 37, 39, 39, 39, 40, 41, 39, 39, 39, 41,
    40]
y = [1, 6, 11, 16, 21, 26, 31, 36, 41]

ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer = True))


color = ['r', 'g', 'y', 'orange', 'crimson', 'k', 'r', 'g']
n, bins, patches = plt.hist(x, y, color = 'green', stacked = True)

for i in range(len(n)):
    patches[i].set_fc(color[i])

    
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Population Age Count')

plt.xticks(y)
plt.yticks()
plt.savefig('histogram')
plt.show()
```

```python
x = np.random.randn(1000, 16)
plt.hist(x, 30, ec = 'k', stacked = True)

plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Population Age Count')
plt.show()
```

```python
x = np.random.randn(1000)
y = np.sin(x)
z = np.cos(x)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Population Age Count', color = 'red', fontsize = 20)

plt.hist([x, y, z], bins = 'auto', ec = 'k', orientation = 'vertical', align = 'mid', log = True, histtype = 'stepfilled')
plt.show()
```
