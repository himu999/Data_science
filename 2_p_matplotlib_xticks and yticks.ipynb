# xicks and yticks 

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
x = np.array([1, 2, 3, 4])

plt.plot(x**2, x, 'ro:', markersize = 8, markeredgecolor = 'k', markerfacecolor = 'k', label = 'squares')
plt.plot(x**2.5, x, 'ko:', markersize = 8, markeredgecolor = 'r', markerfacecolor = 'r', label = 'cubes')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple plot')

plt.ylim(0, 5)
# plt.xlim(0, 40)

plt.xticks(np.arange(0, 41, step = 2), color = 'r', rotation = 'vertical')
plt.yticks(np.arange(0, 5+1), ['A', 'B', 'c', 'D', 'E', 'F'], color = 'r')


plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')


plt.legend(loc = 'upper left', facecolor = 'y', edgecolor = 'r', title= 'legend', fontsize = 12,
           title_fontsize = 13, frameon= True, shadow = True, borderpad = True, framealpha = 0.2, labelspacing = 1)

plt.show()
```

#### Adjusting the ticks properties

```python
x = np.array([1, 2, 3, 4])

ax = plt.
plt.plot(x**2, x, 'ro:', markersize = 8, markerfacecolor = 'k', markeredgecolor = 'k', label = 'label1')
plt.plot(x**2.5, x, 'ko:', markersize = 8, markerfacecolor = 'r', markeredgecolor = 'r', label = 'label2')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple plot')

plt.minorticks_on()
plt.tick_params(length =10, width = 2, color = 'r', direction = 'inout', 
                right =  True, left = True, top = True, pad = 10)
plt.show()
```

```python
x = np.array([1, 2, 3, 4])

ax = plt.figure().gca()
ax.yaxis.set_major_locator(MaxNLocator(integer = True))


plt.plot(x**2, x, 'ro:', markersize = '8', markerfacecolor = 'k', markeredgecolor = 'k', label = 'label1')
plt.plot(x**3, x, 'ko:', markersize = '8', markerfacecolor = 'r', markeredgecolor = 'r', label = 'label2')

plt.ylim(0, 5)


plt.minorticks_on()
plt.grid(b = True, which = 'major', color = 'b')
plt.grid(b = True, which = 'minor')

plt.xticks(rotation = 'vertical', color = 'k', backgroundcolor = 'orange')
plt.yticks(backgroundcolor = 'orange')


plt.tick_params(labelcolor = 'black', direction='inout', right = True, top =  True,  width = 10, length = 10, color = 'r', 
               labelrotation = 120, labelright = True, labeltop = True, which = 'major', grid_color = 'purple')

plt.legend(loc = 'upper left', title= 'Title', fontsize = '12', markerfirst = True, facecolor = 'white', edgecolor = 'red',
          shadow = True, frameon = True, framealpha = 0.02)


plt.show()
```
